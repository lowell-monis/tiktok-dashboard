# Extracting the data

This notebook utilizes `kagglehub` to extract the data set used for this project from Kaggle. Alternatively, you can procure the data directly from [the website](https://www.kaggle.com/datasets/raminhuseyn/dataset-from-tiktok/data).

To avail this notebook, you will need an API secret key from Kaggle saved in an environment file (.env) in this repository. An example of how the contents of the .env file will look like is as follows:

```bash
KAGGLE_API_TOKEN=xxxxxx
```

To get the key, go to your Settings on Kaggle after logging in, scroll till you see API, and generate a new token. Copy the key; this is the first and last time the website will show you the key.

````python
import os
from dotenv import load_dotenv

load_dotenv()

api_key = os.getenv("KAGGLE_API_TOKEN")

if api_key:
    print("API Key loaded successfully!")
    print("Copy the key and use it to login in the next cell.")
    print(api_key)
else:
    print("Failed to load API Key.")
````

Now, run the following cell to setup the login portal. Paste the key when prompted.

````python
import kagglehub
kagglehub.login()
````

Next, we can proceed with downloading and storing the data for your personal use. The repository ignores all data files, so when you clone it, you will not be able to run any code since the data won't be here.

````python
import shutil
import os
import zipfile

download_path = kagglehub.dataset_download("raminhuseyn/dataset-from-tiktok")

files = os.listdir(download_path)
print(f"Files found in download: {files}")

for file in files:
    full_path = os.path.join(download_path, file)
    
    if zipfile.is_zipfile(full_path):
        with zipfile.ZipFile(full_path, 'r') as zip_ref:
            zip_ref.extractall("../data/")
    
    elif file.endswith('.csv'):
        shutil.copy(full_path, "../data/tiktok_dataset.csv")
````
